```python
# PREGUNTA 2
# PREGUNTA 2  (b) Usando matplotlib y/o seaborn grafique un diagrama de barras con los
#tipos de publicación versus la cantidad.
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Cargar el archivo CSV
df = pd.read_csv('scopus.csv')

# Verificar las primeras filas del DataFrame para confirmar la estructura
print(df.head())

# Contar las ocurrencias de cada tipo de publicación
document_type_counts = df['Document Type'].value_counts()

# Crear el gráfico de barras con coloración por tipo de publicación
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=document_type_counts.index, y=document_type_counts.values, palette='viridis')

# Agregar etiquetas y título
plt.xlabel('Tipo de Publicación')
plt.ylabel('Cantidad')
plt.title('Cantidad de Publicaciones por Tipo')

# Formatear el eje y con un formato de número entero
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))

# Rotar las etiquetas del eje x para mejorar la legibilidad
plt.xticks(rotation=45, ha='right')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# PREGUNTA 2
# PREGUNTA 2  (c) - Usando Missingno evalúe si existen datos faltantes
import missingno as msno
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('scopus.csv')

# Mostrar un resumen básico de los datos, incluyendo la cantidad de valores no nulos
print(df.info())

# Visualizar un mapa de calor de los datos faltantes
plt.figure(figsize=(12, 6))
msno.heatmap(df)
plt.title('Mapa de Calor de Datos Faltantes')
plt.show()

# Visualizar un gráfico de barras de datos faltantes por columna
plt.figure(figsize=(12, 6))
msno.bar(df)
plt.title('Gráfico de Barras de Datos Faltantes')
plt.show()

# Visualizar un gráfico de matriz de datos faltantes
plt.figure(figsize=(12, 6))
msno.matrix(df)
plt.title('Matriz de Datos Faltantes')
plt.show()
```
